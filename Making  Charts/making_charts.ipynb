```python
from matplotlib import pyplot as plt

import numpy as np
import pandas as pd
```

```python
def get_last_20000_matching_lines(file_path):
    # 打开文件并读取所有行
    with open(file_path, 'r') as file:
        lines = file.readlines()

    # 筛选符合条件的行
    matching_lines = []
    for line in lines:
        if line.strip(): 
            # 对于QLearning,直接读取文件即为分数 文件名是score.txt
            #matching_lines.append(float(line))
            
            # 对于另外两个，需要做一定的处理，需要读取对应的training_log.txt
            # 检查行是否包含 "Epoch" 和 "score"
            if "Epoch" in line and "score" in line:
                matching_lines.append(float(line.split("score =")[-1].strip()))
            
    
    # 获取倒数20000行
    last_20000_matching_lines = matching_lines[-3000:]
    
    return last_20000_matching_lines


# 调用函数并打印结果
file_path = '##<Your Text File Path>.txt'  # 这里填入文件路径
matching_lines = get_last_20000_matching_lines(file_path)
```

```python
len(matching_lines)
```

```python
data = matching_lines
max_value = np.max(data)         # 最大值
mean_value = np.mean(data)       # 平均值
variance_value = np.var(data)    # 方差

print(f"最大值: {max_value}")
print(f"平均值: {mean_value}")
print(f"方差: {variance_value}")
```

```python
import matplotlib.pyplot as plt
def plot_colored_scatter(scores):
    # 创建一个横坐标的索引列表
    x = list(range(1, len(scores) + 1))  # 从1开始的索引，表示每个分数的序号

    # 创建散点图，`c` 参数用来表示每个点的颜色，`cmap` 指定颜色映射
    plt.figure(figsize=(12, 6))
    scatter = plt.scatter(x, scores, c=scores, cmap='plasma_r', alpha=0.5, edgecolors='w', linewidth=0.5)
    
    plt.ylabel("Score")

    # 添加颜色条
    plt.colorbar(scatter, label='Score')  # 显示颜色条，表示分数的颜色映射
    plt.savefig('scatter_pretrain.png', format='png')  # 保存为 PNG 格式的文件
    # 显示图形
    plt.show()
# 调用绘图函数
plot_colored_scatter(matching_lines)
```
